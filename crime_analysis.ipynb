##### The dataset entries are inconsistent and has capital letters in some columns and small in others. Hence, apart from the columns with abbreviations, all the entries are changed to correct case to ensure uniformity.

```python
cols = ['Crime Name1', 'Crime Name2', 'Crime Name3',
       'Police District Name', 'City', 'State', 
       'Agency', 'Beat', 'PRA',
       'Street Name']
df[cols] = df[cols].apply(lambda x: x.str.capitalize()) #Using the string function to capitalize only the first letter of each entry.
```

It was decided to remove 15% of rows to ensure that only relevant data is retained since more than 15% of the data was missing from dispatch date time. However, we opted to keep the rows intact.
